```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report


import seaborn as sb

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings('ignore')
```

```python
playlistDF = pd.read_csv("cleaned_data.csv")
playlistDF.head()
```

```python
playlistDF.shape
```

```python
playlistDF.info()
```

```python
playlistDF.dropna(inplace=True)
playlistDF.head()
```

```python
playlistDF.info()
```

```python
playlistDF = playlistDF[['artists','track_name', 'popularity', 'track_genre', 'explicit', 'danceability']]
playlistDF.head()
```

```python
playlistDF.nunique()
```

```python
# apply the standard scaler to the numeric columns
playlistScaledDF = StandardScaler().fit_transform(playlistDF[["popularity", "danceability"]])
playlistScaledDF = pd.DataFrame(playlistScaledDF, columns=["popularity", "danceability"])
playlistScaledDF.head()
```

```python
# one hot encode track genre column
playlistFinalDF = pd.get_dummies(playlistDF["track_genre"])
playlistFinalDF.head()
```

```python
# one hot encode explicit column
explicitDF = pd.get_dummies(playlistDF["explicit"])
explicitDF.head()
```

```python
explicitDF.columns
```

```python
#
playlistFinalDF["track_name"] = playlistDF["track_name"]
playlistFinalDF["Not Explicit"] = explicitDF[False]
playlistFinalDF["Explicit"] = explicitDF[True]
playlistFinalDF["Popularity"] = playlistScaledDF["popularity"]
playlistFinalDF["Danceability"] = playlistScaledDF["danceability"]
playlistFinalDF.head()
```

```python
# split the data into X data and y data
X = playlistFinalDF.copy()
X.drop("track_name", axis=1, inplace=True)
X.head()
```

```python
y = playlistFinalDF['track_name'].values.reshape(-1,1)
y[:5]
```

```python
# split the data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# make the instance of the standard scaler
scaler = StandardScaler()
```

```python
# fit the scaler onto the data
X_scaler = scaler.fit(x_train)
```

```python
# scale the x train and test using the transform function
xTrainScaled = X_scaler.transform(x_train)
xTestScaled = X_scaler.transform(x_test)
```

```python
# use the sklearn tree module to make the tree classifier
from sklearn import tree
```

```python
# use .DecisionTreeClassifier from the tree module to make the tree classifier
treeModel = tree.DecisionTreeClassifier()
```

```python
# fit the model onto our scaled X training data and the trained y data
treeModel = treeModel.fit(xTrainScaled, y_train)
```

```python
# make the prediction using the testing data
predictedValues = treeModel.predict(xTestScaled)
```
