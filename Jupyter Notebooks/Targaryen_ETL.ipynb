```python
import pandas as pd
import numpy as np
import os
```

```python
spotify_df = pd.read_csv(os.path.join('..','Resouces','genres_v2.csv'), delimiter=',', low_memory=False) 
long_df = pd.read_csv(os.path.join('..','Resouces','dataset-long.csv'), delimiter=',', low_memory=False, index_col=0)
```

```python
filter_df = long_df.groupby('track_genre').mean('popularity').sort_values('popularity', ascending=False).head(10).index.tolist()
```

```python
long_df_filtered = long_df[long_df['track_genre'].isin(filter_df)].reset_index(drop=True)
```
