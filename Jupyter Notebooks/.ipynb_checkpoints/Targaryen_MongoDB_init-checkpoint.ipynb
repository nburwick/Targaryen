```python
%%time
not_found = 0
for index, track in tqdm(df.iterrows(), total=df.shape[0]):
    time.sleep(1)
    try:
        track_name = track['track_name']
        artist = track['artists']
        result = spotify.search(f'track:{track_name} artist:{artist}')
        # pprint(result)
        url = result['tracks']['items'][0]['external_urls']['spotify']
        track_id = result['tracks']['items'][0]['id']
        # print(url)
        df.loc[index,'track_url'] = url
        df.loc[index, 'track_id']
    except:
        not_found += 1
        df.loc[index,'track_url'] = None
    
```
